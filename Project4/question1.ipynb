```python
import re
actor_movies = []
with open('./project4_data/mergeact_movies.txt', 'rU', encoding = 'ISO-8859-1') as fid :
    for line in fid:
        actor_movies.append(re.split(r'\s*\t\t',line.strip()))
actor_movies
    
          
        
```

```python
actor_movieNo_ = []
for actor_movie in actor_movies:
    name = actor_movie[0]
    number = len(actor_movie)-1
    actor_movieNo_.append([name, str(number)])

actor_movieNo_
```
