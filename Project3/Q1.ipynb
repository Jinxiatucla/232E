## Q1 Heat maps for Reward function 1 and Reward function 2

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(11)
y = np.arange(11)
X, Y = np.meshgrid(x, y)
```

```python
reward_1 = np.zeros((10,10))
```

```python
plt.pcolor(X, Y, reward_1, edgecolors='k', linewidths=1 )
plt.colorbar()
plt.gca().invert_yaxis()
```

```python
reward_2 = np.zeros((10,10))
reward_2[1,4:7] = -100
reward_2[2:7,4] = -100
reward_2[2:4,6] = -100
reward_2[3,7] = -100
reward_2[3:8,8] = -100
reward_2[7,6:8] = -100
reward_2[8,6] = -100
```

```python
plt.pcolor(X, Y, reward_2, edgecolors='k', linewidths=1 )
plt.colorbar()
plt.gca().invert_yaxis()
```
